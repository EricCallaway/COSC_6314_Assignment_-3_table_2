```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, roc_curve, roc_auc_score
```

```python
data_path = '/content/drive/MyDrive/Colab Notebooks/Data Mining/data/assignment_#3_table_2.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
df['Value'] = df['Actual']
```

```python
df['Value'] = df['Value'].map({'Yes': 1, 'No': 0})
```

```python
y_true = df['Actual']
class_f = df['Classifier F']
class_g = df['Classifier G']
class_i = df['Classifier I']
class_j = df['Classifier J']
value = df['Value']
```

Class F's Classification Report

```python
print(classification_report(y_true=y_true, y_pred=class_f))
```

Class G's Classification Report

```python
print(classification_report(y_true=y_true, y_pred=class_g))
```

Class I's Classification Report

```python
print(classification_report(y_true=y_true, y_pred=class_i))
```

```python
print(classification_report(y_true=y_true, y_pred=class_j))
```

```python
df['Class F Binary'] = df['Classifier F'].map({'Yes' : 1, 'No' : 0})
df['Class G Binary'] = df['Classifier G'].map({'Yes' : 1, 'No' : 0})
df['Class H Binary'] = df['Classifier H'].map({'Yes' : 1, 'No' : 0})
df['Class I Binary'] = df['Classifier I'].map({'Yes' : 1, 'No' : 0})
df['Class J Binary'] = df['Classifier J'].map({'Yes' : 1, 'No' : 0})
```

```python
class_f_bin = df['Class F Binary']
class_g_bin = df['Class G Binary']
class_h_bin = df['Class H Binary']
class_i_bin = df['Class I Binary']
class_j_bin = df['Class J Binary']
```

```python
class_f_fpr,class_f_tpr, class_f_threshold = roc_curve(value,class_f_bin)
class_f_r_auc = roc_auc_score(value, class_f_bin)

class_g_fpr,class_g_tpr, class_g_threshold = roc_curve(value,class_g_bin)
class_g_r_auc = roc_auc_score(value, class_g_bin)

class_h_fpr,class_h_tpr, class_h_threshold = roc_curve(value,class_h_bin)
class_h_r_auc = roc_auc_score(value, class_h_bin)

class_i_fpr,class_i_tpr, class_i_threshold = roc_curve(value,class_i_bin)
class_i_r_auc = roc_auc_score(value, class_i_bin)

class_j_fpr,class_j_tpr, class_j_threshold = roc_curve(value,class_j_bin)
class_j_r_auc = roc_auc_score(value, class_j_bin)
```

```python
plt.figure(figsize=(10,10))

plt.plot(class_f_fpr, class_f_tpr, linestyle='--', label='Classifier F Predictions (AUROC = %0.3f)' % class_f_r_auc)
plt.plot(class_g_fpr, class_g_tpr, marker='.', label='Classifier G Predictions (AUROC = %0.3f)' % class_g_r_auc)
plt.plot(class_h_fpr, class_h_tpr, marker='.', label='Classifier H Predictions (AUROC = %0.3f)' % class_h_r_auc)
plt.plot(class_i_fpr, class_i_tpr, marker='.', label='Classifier I Predictions (AUROC = %0.3f)' % class_i_r_auc)
plt.plot(class_j_fpr, class_j_tpr, marker='.', label='Classifier J Predictions (AUROC = %0.3f)' % class_j_r_auc)

plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
